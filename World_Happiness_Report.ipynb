**Happiest** **and** **Unhappiest** **Countries** **in** **2021**

```python
#Happiest and Unhappiest Countries in 2021
print("Unhappiest Countries Score",data_2021["Ladder score"].min())#Unhappiest Countries Score
print("Happiest Countries Score",data_2021["Ladder score"].max())#Happiest Countries Score
print("Average Happiness Score",data_2021["Ladder score"].mean())#Average Happiness Score
```

```python
#Happiest countries of 2021
average_countries_score = data_2021["Ladder score"].mean()
data_2021_happiest = data_2021[data_2021.loc[:,"Ladder score"]>average_countries_score][:5]
print(data_2021_happiest)
print(average_countries_score)
```

```python
#Unhappiest countries of 2021
average_countries_score = data_2021["Ladder score"].mean()
data_2021_unhappiest = data_2021[data_2021.loc[:,"Ladder score"]<average_countries_score].sort_values(by="Ladder score",ascending=True)[:5][::-1]
print(data_2021_unhappiest)
```
